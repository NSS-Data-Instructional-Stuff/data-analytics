```python
police_calls = pd.read_csv('data/Metro_Nashville_Police_Department_Calls_for_Service__2018_.csv')
police_calls.head()
```

```python
police_calls = police_calls.dropna(how = 'any')
```

```python
police_calls['geometry'] = police_calls.apply(lambda x: Point((float(x.Longitude), float(x.Latitude))), axis=1)
police_calls.head()
```

```python
police_calls_geo = gpd.GeoDataFrame(police_calls, crs = zips.crs, geometry = police_calls.geometry)
police_calls_geo.head()
```

```python
zips_small.head()
```

```python
police_calls_with_zip = gpd.sjoin(police_calls_geo, zips_small, op = 'within')
```

```python
police_calls_with_zip.head()
```

```python
police_calls_with_zip.to_csv('police_calls_with_zipcode.csv', index = False)
```
