## Imports go up here

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Do the initial request..

```python
website_url = 'https://en.wikipedia.org/wiki/List_of_unusual_deaths'
result = requests.get(website_url)

result.status_code
```

```python
result.content
```

## Make some soup!

```python
soup = BeautifulSoup(result.content, 'lxml')
print(soup.title)
```

```python
print(soup.meta)
```

### Show the first image

```python
from IPython.core.display import HTML, display

first_image, *remaining_images = images = soup.find_all('img')
display(HTML(str(first_image)))
```

```python
def col_3(image):
    # Creates some html..
    return f'<div class="col-md-3">{str(image)}</div>'


def show(html):
    # displays html
    return display(HTML(html))


def print_values(bundles, by=4):
    # creates rows of images in groups of 4 or less!
    row_set = []
    
    for image in bundles:
        row_set.append(col_3(image))
        
        if len(row_set) == by:
            
            yield show(''.join(row_set))
            row_set = []
            
    yield show(''.join(row_set))
```

```python
[row for row in print_values(remaining_images)];
```
